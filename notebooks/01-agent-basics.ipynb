# 01 - Agent Basics

A minimal notebook to verify the LangChain/LangGraph stack works. Loads `.env` from the project root and runs a simple ChatOpenAI call.

```python
import os
from dotenv import load_dotenv

# Load .env from project root (dotenv searches cwd and parent dirs)
load_dotenv()

assert os.getenv("OPENAI_API_KEY"), "OPENAI_API_KEY must be set in .env at project root"
print("OPENAI_API_KEY is set. Ready to run.")
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage

model = ChatOpenAI(model="gpt-4o-mini", temperature=0.3, max_tokens=500)
response = model.invoke([HumanMessage(content="What are macros? Answer in one sentence.")])
print(response.content)
```
